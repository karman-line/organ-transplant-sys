## Data Cleaning! ##

**Objective:** Get a .csv file with each OPO, its city, and geographical coordinates in order to map it on Datawrapper


**Files:** 1. us_cities_loc.csv (obtained from https://github.com/kelvins/US-Cities-Database)
2. opo_list.xlsx (Extracted from https://www.srtr.org/reports/opo-specific-reports/)

```python
import pandas as pd
import numpy as np
```

```python
cities = pd.read_csv('/Users/karthikanamboothiri/Desktop/Data/us_cities_loc.csv')
cities.head()
```

```python
opo = pd.read_excel('opo_list.xlsx', engine='openpyxl')
opo.head()
```

```python
opo = opo.rename(columns={'State (Abb)': 'STATE_CODE', 'City': 'CITY'})
opo.head()
           
```

```python
opo_loc = opo.merge(cities, how='inner', left_on=["CITY", "STATE_CODE"], right_on=["CITY", "STATE_CODE"])
opo_loc
```

```python
#Keeping just the first entry
df = opo_loc.drop_duplicates(subset='Name', keep="first")
df.head()
```

```python
df.shape
```

```python
df.to_csv('opo_loc.csv', index=False)
```
